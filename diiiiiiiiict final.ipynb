```python
import pandas as pd

#Reads annotations and true skill ratings and creates a dictionary of len 4000 comprising only the images 
# with true skill ratings in my_dict
my_dict={}
with open('C:/Users/luca.chech.16/Desktop/Visual Search Exp/annotations.csv', 'r') as f:
    for line in f:
        items = line.split(',')
        key, values = items[0], items[1:]
        my_dict.setdefault(key, []).extend(values)
lista_all= my_dict.keys()

true_skill_dict = {}
with open('C:/Users/luca.chech.16/Desktop/Visual Search Exp/true_skill.csv', 'r') as f:
    for line in f:
        items = line.split(',')
        key, values = items[0], items[1:]
        true_skill_dict.setdefault(key, []).extend(values)
lista_subset = true_skill_dict.keys()

for each in my_dict:
    if each in true_skill_dict:
        my_dict[each].append(float(true_skill_dict[each][0]))
my_dict
keys_to_delete = []
for each in my_dict:
    if each not in true_skill_dict:
        keys_to_delete.append(each)

for each in keys_to_delete:
    del my_dict[each]
my_dict

for each in my_dict:
    if my_dict[each][-1] <= 21.26:
        my_dict[each].append('low')
    elif my_dict[each][-1] >= 28.62:
        my_dict[each].append('high')
    else:
        my_dict[each].append('medium')
df = pd.DataFrame(my_dict) 
df = df.T
df = df[df[7] != 'medium']
print df
df.to_csv('my_dict.csv')
```
