```python
import pandas as pd
```

```python
df = pd.DataFrame.from_csv('C:/Users/nilli lab/Desktop/load/image_selection_criteria/my_dict.csv')
df.reset_index(drop=False, inplace=True)
#df.reset_index(level=None, drop=False, inplace=False, col_level=0, col_fill='')
cols = ['image_name','cat1','cat2','cat3','cat4','cat5','cat6','ts_rating','load']
df.columns = cols
print df['ts_rating'].quantile([.25, .75])
print df.shape
df
```

```python
for i in range(df.shape[0]):
    if df.loc[i,'ts_rating'] < 18.65:
        df.loc[i,'load'] = 'low'
    elif df.loc[i,'ts_rating'] > 31.18:
        df.loc[i,'load'] = 'high'
    else:
        df.loc[i,'load'] = 'intermediate'
        
```

```python
df_low = df[df['load']=='low']
df_high = df[df['load']=='high']
```

```python
cat1 = df['cat1'].tolist()
cat2 = df['cat2'].tolist()
cat3 = df['cat3'].tolist()
cat4 = df['cat4'].tolist()
cat5 = df['cat5'].tolist()
cat6 = df['cat6'].tolist()
load = df['load'].tolist()
ts_rating = df['ts_rating'].tolist()
image_name = df['image_name'].tolist()

images_dict ={z[0]:list(z[1:]) for z in zip(image_name,cat1,cat2,cat3,cat4,cat5,cat6,load,ts_rating)}
```

```python
categories = [' bottle', ' horse', ' pottedplant', ' dog',
              ' cat', ' person', ' aeroplane', ' car', ' chair', ' sofa', ' bird', ' boat']
```

```python
categories_counters = {}
categories_counters[' bottle'] = ['low',0,'high',0]
categories_counters[' horse'] = ['low',0,'high',0]
categories_counters[' pottedplant'] = ['low',0,'high',0]
categories_counters[' dog'] = ['low',0,'high',0]
categories_counters[' cat'] = ['low',0,'high',0]
categories_counters[' person'] = ['low',0,'high',0]
categories_counters[' aeroplane'] = ['low',0,'high',0]
categories_counters[' car'] = ['low',0,'high',0]
categories_counters[' chair'] = ['low',0,'high',0]
categories_counters[' sofa'] = ['low',0,'high',0]
categories_counters[' bird'] = ['low',0,'high',0]
categories_counters[' boat'] = ['low',0,'high',0]
print len(categories_counters.keys())
```

```python

for i in categories_counters:
    for i_2 in images_dict:
        if i in images_dict[i_2] and 'low' in images_dict[i_2]:
            categories_counters[i][1] = categories_counters[i][1] +1
        elif i in images_dict[i_2] and 'high' in images_dict[i_2]:
            categories_counters[i][3] = categories_counters[i][3] +1
```
