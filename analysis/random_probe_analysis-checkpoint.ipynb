```python
# Loading libraries
import pandas as pd
import numpy as np

# Reading csv file and changing column names
mypath = 'C:/Users/luca.chech.16/Desktop/Random_probes_last/Luca/'
df_1 = pd.DataFrame.from_csv(mypath + participant +'_experiment_ordered.csv')
df_1.columns = ['Image_Name','Cat_1','Cat_2','Cat_3','Cat_4','Cat_5','True_Skill_Rating','Load','Exemplar','To_Be_Asked','Trial_Type',
                'Audio','RT_TO_1','RT_VS','RT_TO_2','Sound_Onset', 'Sound_Hz','Trial_N_Block','Trial_N' ]

#Converting useful dataframe columns into lists
RT_TO_1 = df_1['RT_TO_1'].tolist()
RT_TO_2 = df_1['RT_TO_2'].tolist()
RT_VS = df_1['RT_VS'].tolist()
Trial_Type = df_1['Trial_Type'].tolist()
Audio = df_1['Audio'].tolist()
Load = df_1['Load'].tolist()
Trial_N = df_1['Trial_N'].tolist()
Sound_Onset = df_1['Sound_Onset'].tolist()
```

```python
RT_TO_1[46] = 'None'
#%pdb
```

```python
#######
rt_to_cleaned_i = []
rt_to_cleaned = []
for i in [i for i,x in enumerate(RT_TO_1) if len(x) != 4]:
    #print i
    rt_to_cleaned_i.append(i)
    
for each in RT_TO_1:
    if len(each) != 4:
        #print each[11:16]
        rt_to_cleaned.append(each[11:16])

dictionary1 = dict(zip(rt_to_cleaned_i, rt_to_cleaned))
#print dictionary1
#######
```

```python
#######
rt_to_cleaned_2_i = []
rt_to_cleaned_2 = []

for i in [i for i,x in enumerate(RT_TO_2) if len(x) >4]:
    #print i
    rt_to_cleaned_2_i.append(i)
    
for each in RT_TO_2:
    if len(each) >4:
        #print each[10:15]
        rt_to_cleaned_2.append(each[10:15])

dictionary2 = dict(zip(rt_to_cleaned_2_i, rt_to_cleaned_2))
#######
```

```python
#######
RT_TO = []
for i in range(480):
    if i in dictionary1:
        RT_TO.append(dictionary1[i])
    elif i in dictionary2:
        RT_TO.append(dictionary2[i])
    else:
        RT_TO.append('NaN')
#print RT_TO


#print len(RT_TO)
#######
```

```python
#######
RT_VS_cleaned = []
response_key=[]

for each in RT_VS:
    if len(each) != 2:
        RT_VS_cleaned.append(each[7:12])
        response_key.append(each[3:4])
    elif len(each) == 2:
        RT_VS_cleaned.append('NaN')
        response_key.append('NaN')
#print len(RT_VS_cleaned)
#######
```

```python
RT_TO = [float(n) if n != 'NaN' else n for n in RT_TO]
```

```python
RT_VS_cleaned = [float(n)-1 if n != 'NaN' else n for n in RT_VS_cleaned]
```

```python
Sound_Onset = [float(n) if n != 'None' else n for n in Sound_Onset]
```

```python
for i in range(0,480):
    if type(RT_TO[i]) == float and type(Sound_Onset[i]) == float and RT_TO[i] < Sound_Onset[i]:
        #print i, RT_TO[i], Sound_Onset[i]
        RT_TO[i] = RT_TO[i] +1
```

```python
for i in range(0,480):
    if type(RT_TO[i]) == float and type(Sound_Onset[i]) == float:
        #print i, RT_TO[i], Sound_Onset[i]
        RT_TO[i] = RT_TO[i] - Sound_Onset[i]
```

```python
counter_TO = 0
RT_TO_floats = [n for n in RT_TO if type(n)==float]
sd_TO = np.std(RT_TO_floats)
mean_TO = np.mean(RT_TO_floats)
RT_TO_no_out = []
for i in RT_TO:
    if type(i) != float:
        RT_TO_no_out.append(i)
    elif type(i) == float:
        if (mean_TO - 2.5 * sd_TO) < i < (mean_TO + 2.5 * sd_TO):
            RT_TO_no_out.append(i)
        else:
            counter_TO = counter_TO + 1
            print 'Outlier ' + str(counter_TO) + ': ' + str(i) 
            RT_TO_no_out.append(mean_TO)
RT_TO = RT_TO_no_out

RT_TO = [float(n) if type(n) != float and type(n) != str else n for n in RT_TO]
```

```python
print len(RT_VS_cleaned)
counter_VS = 0
RT_VS_floats = [n for n in RT_VS_cleaned if type(n)==float]
sd_VS = np.std(RT_VS_floats)
mean_VS = float(np.mean(RT_VS_floats))
RT_VS_no_out = []
for i in RT_VS_cleaned:
    if type(i) != float:
        RT_VS_no_out.append(i)
    elif type(i) == float:
        if (mean_VS - 2.5 * sd_VS) <= i <= (mean_VS + 2.5 * sd_VS):
            RT_VS_no_out.append(i)
        else:
            counter_VS = counter_VS + 1
            print 'Outlier ' + str(counter_VS) + ': ' + str(i) 
            RT_VS_no_out.append(mean_VS)
RT_VS_cleaned = RT_VS_no_out
RT_VS_cleaned = [float(n) if type(n)  != str else n for n in RT_VS_cleaned]
```

```python
#######
Acc_VS = []
for i in range(0,480):
    if 'Critical' in Audio[i]:
        Acc_VS.append('999')
    elif 'Normal' in Audio[i] and RT_TO[i] != 'NaN':
        Acc_VS.append('666')
    elif Trial_Type[i] == 'Target Present' and response_key[i] == 'p':
        Acc_VS.append(1)
    elif Trial_Type[i] == 'Target Present' and response_key[i] == 'q':
        Acc_VS.append(0)
    elif Trial_Type[i] == 'Target Absent' and response_key[i] == 'q':
        Acc_VS.append(1)   
    elif Trial_Type[i] == 'Target Absent' and response_key[i] == 'p':
        Acc_VS.append(0)
    elif Trial_Type[i] == 'No Target':
        Acc_VS.append('999')

#print len(Acc_VS)
#######
```

```python
#######
# CREATING Acc_DT (1 = included for analysis; 0 = excluded from analysis)
Acc_DT = []
for i in range(0,480):
    if Audio[i] == 'Critical' and RT_TO[i] != 'NaN':
        #print i, Acc_VS[i], type(Acc_VS[i])
        if 'Target Present' in Trial_Type[i] and 'p' in response_key[i]:
            Acc_DT.append(1)
        elif 'Target Absent' in Trial_Type[i] and 'q' in response_key[i]:
            Acc_DT.append(1)
        elif 'Target Present' in Trial_Type[i] and 'q' in response_key[i]:
            Acc_DT.append(999)
        elif 'Target Absent' in Trial_Type[i] and 'p' in response_key[i]:
            Acc_DT.append(999)
        elif 'No Target' in Trial_Type[i]:
            Acc_DT.append(1)
    elif Audio[i] == 'Critical' and RT_TO[i] == 'NaN':
        Acc_DT.append(0)
    elif Audio[i] == 'Normal' and RT_TO[i] != 'NaN':
        Acc_DT.append(666)
    elif Audio[i] == 'Normal' and RT_TO[i] == 'NaN':
        Acc_DT.append(999)
   
```
