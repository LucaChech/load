```python
import pandas as pd
import numpy as np
import random
import csv
df = pd.DataFrame.from_csv('C:/Users/luca.chech.16/Desktop/Visual Search Exp/my_dict_final.csv')
df.reset_index(drop=False)
```

```python
image_names = df.index.tolist()
cat1 = df['0'].tolist()
cat2 = df['1'].tolist()
cat3 = df['2'].tolist()
cat4 = df['3'].tolist()
cat5 = df['4'].tolist()
cat6 = df['5'].tolist()
#ts = df['6'].tolist()
load = df['7'].tolist()
my_dict ={z[0]:list(z[1:]) for z in zip(image_names,cat1,cat2,cat3,cat4,cat5,cat6,load)}
```

```python
#categories = [' bottle', ' horse', ' cat', ' person', ' aeroplane']
p_load = ['low','high']
target_present_trials = []
target_absent_trials = []
lista= my_dict.keys()

categories = [' bottle', ' horse', ' pottedplant', ' dog', ' cat', ' person', ' aeroplane', ' car', ' chair', ' sofa', ' bird', ' boat']



```

```python
# Creating a subset of items representing the possible images to be used in the experiment and deleting every other 
# element from my_dict
target_present = []
intermediate = []

for each_1 in categories:
    for load in p_load:
        i = 0
        while True:
            trial = np.random.choice(lista, 1, replace= False)
            a = str(trial)
            a =a[2:-2]
            if each_1 in my_dict[a] and load in my_dict[a] and a not in intermediate and a not in target_present:
                intermediate.append(a)
            if len(intermediate) == 20:
                for each in intermediate:
                    target_present.append(each)
                    #my_dict[each].append(each_1)
                intermediate = []
                break

print len(target_present)

for each in target_present:
    print each
    print my_dict[each]
#[' bottle', ' horse', ' cat', ' person', ' aeroplane']   
```

```python
possible_images_dict = {key:my_dict[key] for key in target_present}
```

```python
# DEFINING TARGET-PRESENT TRIALS
lista= possible_images_dict.keys()
target_present = []
intermediate = []

for each_1 in categories:
    for load in p_load:
        i = 0
        while True:
            trial = np.random.choice(lista, 1, replace= False)
            a = str(trial)
            a =a[2:-2]
            if each_1 in my_dict[a] and load in my_dict[a] and a not in intermediate and a not in target_present:
                intermediate.append(a)
            if len(intermediate) == 10:
                for each in intermediate:
                    target_present.append(each)
                    my_dict[each].append(each_1)
                intermediate = []
                break

print len(target_present)

for each in target_present:
    print each
    print my_dict[each]
```

```python
target_present_dict = {key:possible_images_dict[key] for key in target_present}
target_absent_dict = dict(possible_images_dict)
for each in target_present:
    if each in target_absent_dict:
        del target_absent_dict[each]
exp_dict = dict(target_present_dict)
exp_dict.update(target_absent_dict)
```

```python
for each in target_present_dict:
    target_present_dict[each].append('Target Present')
for each in target_absent_dict:
    target_absent_dict[each].append('None')
    target_absent_dict[each].append('Target Absent')
```
